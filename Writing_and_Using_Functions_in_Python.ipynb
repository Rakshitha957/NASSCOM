```python
#1.DEFINING A BASIC FUNCTION
def greet():
    print("Hello! Welcome to Python Functions.")
# Calling the function
greet()
```

```python
#2.FUNCTION WITH PARAMETERS
def greet_user(name):
    print(f"Hello, {name}!")
greet_user("Alice")
greet_user("Bob")
```

```python
# 3.FUNCTION WITH DEFAULT PARAMETERS
def greet_user(name="Guest"):
    print(f"Welcome, {name}!")
greet_user()
greet_user("Ravi")
```

```python
#4.USING *args (VARIABLE NUMBER OF POSITIONAL ARGUMENTS)
def print_names(*names):
    print("Names passed:", names)
    for name in names:
        print("Hello", name)
print_names("Alice", "bob", "Charlie")
```

```python
#5.USING **kwargs (VARIABLE NUMBER OF KEYWORD ARGUMENTS)
def print_info(**info):
    print("User Info:")
    for key, value in info.items():
        print(f"{key} = {value}")
print_info(name="Alice", age=25, city="Delhi")
```

```python
#6.COMBINING *args and **kwargs
def show_details(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)
show_details(1, 2, 3, name="Ravi", city="Mumbai")
```

```python
# 7. FUNCTION NESTING (CALL ONE FUNCTION INSIDE ANOTHER)
def multiply(a, b):
    return a * b
def square(n):
    return multiply(n, n)
print("Square of 4:", square(4))
```
