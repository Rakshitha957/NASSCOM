```python
# String
c = "Python"
print(type(c))
# Boolean
d = True
print(type(d))
# List
e = [1, 2, 3, "four"]
print(type(e), e)
# Dictionary
f = {"name": "Alice", "age": 25}
print(type(f), f)
```
